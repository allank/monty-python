```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open('offline/transdetails.htm', 'r') as f:
    trans_doc = f.read()
soup = BeautifulSoup(trans_doc, 'html.parser')
```

```python
trans_table = soup.find("table")
```

```python
records = []
for row in trans_table.findAll('tr'):
    col = row.findAll('td')
    if len(col) > 0:
        episodes = ''.join(str(child) for child in col[0].children).replace('</br>', '').replace('<br/>', '').split('<br>')
        series = ''.join(str(child) for child in col[1].children).replace('</br>', '').replace('<br/>', '').split('<br>')
        trans_dates = ''.join(str(child) for child in col[2].children).replace('</br>', '').replace('<br/>', '').split('<br>')
        record_dates = ''.join(str(child) for child in col[4].children).replace('</br>', '').replace('<br/>', '').split('<br>')
        for e, s, t, r in zip(episodes, series, trans_dates, record_dates):
            records.append({
                'episode': e.strip(),
                'series': s.strip(),
                'transmission_date': t.strip(),
                'record_date': r.strip()
            })
```

```python
df = pd.DataFrame(records)
df['transmission_date'] = pd.to_datetime(df['transmission_date'])
df['record_date'] = pd.to_datetime(df['record_date'])
```
